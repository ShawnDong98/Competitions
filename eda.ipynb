```python
import os
from termcolor import colored
import matplotlib.pyplot as plt 

import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold

from config import config
from engine.trainer.utils import seed_everything

df = pd.read_csv(os.path.join(config.root, 'train.csv')) if not config.debug else pd.read_csv(os.path.join(config.root, 'train.csv'))[:1000]

df['file_path'] = df['Id'].apply(lambda x: os.path.join(config.root, 'train', x + '.jpg'))
df["kfold"] = -1
num_bins = int(np.floor(1 + np.log2(len(df))))
df.loc[:, "bins"] = pd.cut(df["Pawpularity"], bins=num_bins, labels=False)
skf = StratifiedKFold(n_splits=config.n_splits, shuffle=True, random_state=config.seed)
folds = enumerate(skf.split(df['file_path'], df['bins']))
```

# Fold1

```python
fold_1, (train_idx_fold1, val_idx_fold1) = next(folds)
train_df_fold1 = df.loc[train_idx_fold1].reset_index(drop=True)
val_df_fold1 = df.loc[val_idx_fold1].reset_index(drop=True)

# Pawpularity Score Distribution
print(colored('Pawpularity Statistics', 'yellow'))
display(train_df_fold1['Pawpularity'].describe())
print()

plt.figure(figsize=(15,8))
plt.title('Train Data Pawpularity Score Distribution', size=24)
plt.hist(train_df_fold1['Pawpularity'], bins=32)
plt.show()


# Pawpularity Score Distribution
print(colored('Pawpularity Statistics', 'yellow'))
display(val_df_fold1['Pawpularity'].describe())
print()

plt.figure(figsize=(15,8))
plt.title('Train Data Pawpularity Score Distribution', size=24)
plt.hist(val_df_fold1['Pawpularity'], bins=32)
plt.show()
```
