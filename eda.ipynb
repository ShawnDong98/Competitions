```python
import os
import pandas as pd
from config import config

train_df = pd.read_csv(os.path.join(config.root, 'train.csv'), sep='\t')
train_df['fnames'] = train_df['fnames'].apply(lambda x: os.path.join(config.root, 'image', 'train', x))
train_df.head()
```

```python
from dataset import FackFaceDetLoader
import matplotlib.pyplot as  plt
sample_dataloader = FackFaceDetLoader(train_df, train_df, config).train_dataloader()
images, labels = iter(sample_dataloader).next()
plt.figure(figsize=(12, 12))
for it, (image, label) in enumerate(zip(images[:16], labels[:16])):
    plt.subplot(4, 4, it + 1)
    plt.imshow(image.permute(1, 2, 0).numpy())
    print(image.mean())
    plt.title(label)
    plt.axis('off')
```

# 

```python
from tqdm import tqdm
import cv2
widths = []
heights = []
ratios = []
for file_path in tqdm(train_df['fnames']):
    image = cv2.imread(file_path)
    h, w, _ = image.shape
    heights.append(h)
    widths.append(w)
    ratios.append(w / h)
```

# Images Height and Width Distribution

```python
from termcolor import colored
# Images Height and Width Distribution
print(colored('Width Statistics', 'yellow'))
display(pd.Series(widths).describe())
print()

print(colored('Height Statistics', 'yellow'))
display(pd.Series(heights).describe())
print()

plt.figure(figsize=(15,8))
plt.title(f'Images Height and Width Distribution', size=24)
plt.hist(heights, bins=32, label='Image Heights')
plt.hist(widths, bins=32, label='Image Widths')
plt.legend(prop={'size': 16})
plt.show()
```

# Images Ratio Distribution

```python
# Images Ratio Distribution
print(colored('Ratio Statistics', 'yellow'))
display(pd.Series(ratios).describe())
print()

plt.figure(figsize=(15,8))
plt.title(f'Images Ratio Distribution', size=24)
plt.hist(ratios, bins=16, label='Image Heights')
plt.legend(prop={'size': 16})
plt.show()
```
